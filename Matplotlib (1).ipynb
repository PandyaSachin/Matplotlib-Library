# Linear Plot

```python
x=[1,2,3,4]
y=[5,6,7,8]

plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
c='r'
plt.plot(x,y,c)
plt.show()
```

```python
x=[2,5,12,5,7]
y=[7,15,20,9,4]
c='b'
plt.plot(x,y,c)
plt.show()
```

```python
x=[2,5,12,5,7]
y=[7,15,20,9,4]
c='b'
plt.plot(x,y,marker='o')
plt.show()
```

```python
x=[2,5,12,5,7]
y=[7,15,20,9,4]

plt.plot(x,y)
plt.plot(y,marker='<')
plt.plot(x,marker='>')
plt.show()
```

# Bar Graph 

```python
x=[1,2,3,4]
y=[5,6,7,8]
plt.bar(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
c=['r','b','g','y']
plt.bar(x,y,color=c)
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7,8]
y=[3,4,6,7,3,8,2,9]

plt.scatter(x,y)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[3,4,2,6,8,9,2,4]
c=['r','b','g','y','r','b','g','y']
plt.scatter(x,y,color=c)
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[3,4,2,6,8,9,2,4]
c=['r','b','g','y','r','b','g','y']
plt.scatter(x,y,color=c)
plt.title("Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Number")
plt.show()
```

```python
x=[1,2,3,4,5,6,7,8]
y=[3,4,2,6,8,9,2,4]
c=['r','b','g','y','r','b','g','y']
plt.scatter(x,y,color=c,s=150)
plt.title("Scatter Plot")
plt.xlabel("Month")
plt.ylabel("Number")
plt.show()
```

# Perform operation on images

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

fname=r'11.jpg'

#opening image using pil
image=Image.open(fname).convert('L')

#mapping image to gray scale
plt.imshow(image, cmap='gist_earth')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

fname=r'11.jpg'

#opening image using pil
image=Image.open(fname).convert('RGB')

#mapping image to gray scale
plt.imshow(image, cmap='gist_earth')
plt.show()
```

#  Pie Chart 

```python
import matplotlib.pyplot as plt
```

```python
x=[10,40,50,70]
y=['English','Maths','Science','Marathi']

plt.pie(x)
plt.show()
```

```python
x=[10,40,50,70]
y=['English','Maths','Science','Marathi']

plt.pie(x, labels=y)
plt.show()
```

```python
x=[10,40,50,70]
y=['English','Maths','Science','Marathi']
c=['r','g','b','y']
plt.pie(x, labels=y, colors=c)
plt.show()
```

```python
x=[10,40,50,70]
y=['English','Maths','Science','Marathi']
c=['r','g','b','y']
plt.pie(x, labels=y, colors=c)
plt.legend()
plt.show()
```
